```python
import pandas as pd

# Read sleep segment data from file
data = pd.read_json("..\Fit\All Data\derived_com.google.sleep.segment_com.google.an.json")

# Extract sleep segments from data
sleep_segments = data["Data Points"]

# Define a dictionary to map intVals to sleep stages
intval_map = {
    1: "Awake",
    2: "Sleep",
    3: "Out-of-bed",
    4: "Light Sleep",
    5: "Deep Sleep",
    6: "REM"
}

# Create an empty DataFrame to store sleep data
sleep_df = pd.DataFrame(columns=["Date", "Duration", "Awake", "Light Sleep", "Deep Sleep", "REM Sleep"])

# Loop through sleep segments and calculate duration and sleep stage counts
for i in range(len(sleep_segments) - 1):
    start_time = sleep_segments[i]["startTimeNanos"] // 1000000000
    end_time = sleep_segments[i + 1]["startTimeNanos"] // 1000000000
    duration = (end_time - start_time) // 60  # in minutes

    # Get sleep stage counts for segment
    stage_counts = {}
    print(sleep_segments[i])
    for data_set in sleep_segments[i]["value"]:
        intval = data_set["intVal"]
        stage = intval_map[intval]
        stage_counts[stage] = stage_counts.get(stage, 0) + 1

    # Add row to DataFrame
    date = pd.to_datetime(start_time, unit="s").strftime("%Y-%m-%d")
    row = [date, duration] + [stage_counts.get(stage, 0) for stage in ["Awake", "Light Sleep", "Deep Sleep", "REM"]]
    sleep_df.loc[len(sleep_df)] = row

# Convert duration to hours
sleep_df["Duration"] /= 60

# Set date column as index
sleep_df.set_index("Date", inplace=True)
```

```python
import datetime
import json
import pandas as pd

# Read sleep segment data from file
data = pd.read_json("..\Fit\All Data\derived_com.google.sleep.segment_com.google.an.json")
sleep_segments = data["Data Points"]

# Define sleep stage type labels and corresponding values
sleep_stage_labels = {1: "Awake (during sleep cycle)", 2: "Sleep", 3: "Out-of-bed", 4: "Light sleep", 5: "Deep sleep", 6: "REM"}

# Convert Epoch time to datetime objects and adjust dates
for segment in sleep_segments:
    start_time = datetime.datetime.fromtimestamp(int(segment["startTimeNanos"]) / 1e9)
    end_time = datetime.datetime.fromtimestamp(int(segment["endTimeNanos"]) / 1e9)
    if end_time.hour >= 16:
        segment_date = end_time.date()
    else:
        segment_date = start_time.date()
    segment["start_time"] = start_time
    segment["end_time"] = end_time
    segment["date"] = segment_date
    segment["sleep_stage"] = sleep_stage_labels[segment["fitValue"][0]["value"]["intVal"]]

# Group sleep segments by date and sleep stage type and sum durations
totals = {}
for segment in sleep_segments:
    date = segment["date"]
    sleep_stage = segment["sleep_stage"]
    if date not in totals:
        totals[date] = {}
    if sleep_stage not in totals[date]:
        totals[date][sleep_stage] = 0
    duration = (segment["end_time"] - segment["start_time"]).total_seconds() / 3600
    totals[date][sleep_stage] += duration

# Create a pandas DataFrame with date and sleep stage type durations
df = pd.DataFrame.from_dict(totals, orient="index").reset_index().rename(columns={"index": "Date"})
df = pd.melt(df, id_vars=["Date"], value_vars=sleep_stage_labels.values(), var_name="Sleep Stage Type", value_name="Duration (hrs)")

# Add Total Sleep column that sums up the duration of all sleep stages (including awake time)
df["Total Sleep"] = df.groupby("Date")["Duration (hrs)"].transform(sum)

# Add Total column that calculates the total time (in hours) of each day
df["Total"] = 24.0

# Add 24 hours to the total if the date corresponds to the previous day
if df.shape[0] > 1:
    if df.iloc[0]["Date"] == df.iloc[1]["Date"] - datetime.timedelta(days=1):
        df.iloc[0, -2:] += df.iloc[1, -2:]
        df = df.iloc[1:]

# Reset the index of the DataFrame
df = df.reset_index(drop=True)

print(df)
```
